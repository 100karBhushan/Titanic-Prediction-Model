skew() to calculate the skewness of each column. The output will be a Series object that contains the skewness values for each column.

Positive skewness indicates that the data is skewed to the right, meaning that the tail on the right side of the distribution is longer or fatter than the left side. Negative skewness indicates that the data is skewed to the left, with a longer or fatter tail on the left side. A skewness value of 0 suggests a roughly symmetric distribution.

```python
train.skew()
```

```python
train.isnull().sum()
```
